```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

#Data
data = {
    'City': [1, 2, 3, 4, 5, 6, 7, 8],
    'Average Income (1000$)': [5, 10, 20, 8, 4, 6, 12, 15],
    'Pizza Sell (1000 pcs)': [27, 46, 73, 40, 30, 28, 46, 59]
}

#Creating DataFrame
df = pd.DataFrame(data)
```

```python
#Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Average Income (1000$)'], df['Pizza Sell (1000 pcs)'])
plt.title('Scatter plot of Average Income vs Pizza Sales')
plt.xlabel('Average Income (1000$)')
plt.ylabel('Pizza Sell (1000 pcs)')
plt.grid(True)
plt.show()
```

```python
#Correlation
correlation = np.corrcoef(df['Average Income (1000$)'], df['Pizza Sell (1000 pcs)'])[0, 1]
print("Correlation coefficient:", correlation)
```

```python
#Linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Average Income (1000$)'], df['Pizza Sell (1000 pcs)'])
print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)
```

```python
#Calculating SSR, SSE, SST
y_pred = slope * df['Average Income (1000$)'] + intercept
SSR = np.sum((y_pred - np.mean(df['Pizza Sell (1000 pcs)']))**2)
SSE = np.sum((df['Pizza Sell (1000 pcs)'] - y_pred)**2)
SST = SSR + SSE
print("SSR:", SSR)
print("SSE:", SSE)
print("SST:", SST)
```
